```python
import pandas as pd
import os 

home = os.path.expanduser("~")
data_path = home + "/Desktop/R_js/data/"

df_pro = pd.read_csv(data_path + "pro_doc_sampled.csv")
df_con = pd.read_csv(data_path + "con_doc_sampled.csv")

display(df_pro.head())
display(df_con.head())
```

```python
df_pro['Pro_Con'] = "pro"
df_con['Pro_Con'] = "con"

display(df_pro.head())
display(df_con.head())
```

```python
df = pd.concat([df_pro, df_con], axis=0, ignore_index=True)
display(df.head())
display(df.shape)
display(df_pro.shape[0]+df_con.shape[0])
```

```python
df.to_csv(data_path + "combined_sampled.csv")
```
