```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("~/Desktop/R_js/data1/diagnosis/diag_8_16.csv")
```

```python
df.head()
```

```python

f, axes = plt.subplots(3, 2, figsize=(14, 10))


sns.lineplot(  y="exclus", x= "num_topics", data=df, markers=True, ax=axes[0, 0])
sns.lineplot(  y="semcoh", x= "num_topics", data=df, markers="o", ax=axes[0, 1])
sns.lineplot(  y="heldout", x= "num_topics", data=df, markers="o", ax=axes[1, 0])
sns.lineplot(  y="residual", x= "num_topics", data=df, markers="o", ax=axes[1, 1])
sns.lineplot(  y="bound", x= "num_topics", data=df, markers="o", ax=axes[2, 0])
sns.lineplot(  y="lbound", x= "num_topics", data=df, markers="o", ax=axes[2, 1])

plt.show()
```

```python
df1 = pd.read_csv("~/Desktop/R_js/data1/diagnosis/diag_17_20.csv")
df1.head()
```

```python
df_df1 = pd.concat([df, df1], axis= 0)
df_df1.head()
```

```python
df_df1
```

```python
f, axes = plt.subplots(3, 2, figsize=(14, 10))
plt.xticks(np.arange(8, 21, 1))

sns.lineplot(  y="exclus", x= "num_topics", data=df_df1, markers=True, ax=axes[0, 0])

sns.lineplot(  y="semcoh", x= "num_topics", data=df_df1, markers=True, ax=axes[0, 1])

sns.lineplot(  y="heldout", x= "num_topics", data=df_df1, markers=True, ax=axes[1, 0])

sns.lineplot(  y="residual", x= "num_topics", data=df_df1, markers=True, ax=axes[1, 1])

sns.lineplot(  y="bound", x= "num_topics", data=df_df1, markers=True, ax=axes[2, 0])

sns.lineplot(  y="lbound", x= "num_topics", data=df_df1, markers=True, ax=axes[2, 1])

plt.show()
```

```python
eval_metrics = ["exclus", "semcoh", "heldout", "residual", "bound", "lbound"]
f, ax = plt.subplots(3, 2, figsize=(14, 10))

for metric in eval_metrics: 
    ax[0, 0].plot(df_df1.num_topics, df_df1.exclus, marker="o")
    ax[0, 0].set_xticks(np.arange(8, 21, 1))
    ax[0, 0].set_xlabel("Num Topics")
    ax[0, 0].set_ylabel("Semantic Coherence")

plt.suptitle("Diagnostic Plot (Combo-STM)")
plt.show()
```
