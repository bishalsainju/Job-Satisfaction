```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv("~/Desktop/R_js/data1/diagnosis/con_diag_4_20.csv")
```

```python
df.head()
```

```python
eval_metrics = ["exclus", "semcoh", "heldout", "residual", "bound", "lbound"]
f, ax = plt.subplots(3, 2, figsize=(18, 12))

i = 0
j = 0
for metric in eval_metrics: 
    print(metric)
    ax[i, j].plot(df["num_topics"], df[metric], marker="o")
    ax[i, j].set_xticks(np.arange(4, 21, 1))
    ax[i, j].set_xlabel("num_topic")
    ax[i, j].set_ylabel(metric)
    if (j == 1): 
        i = (i+1) % 3
    j = (j+1) % 2

plt.suptitle("Diagnostic Plot (Pro-STM)")

root = os.path.expanduser("~")
plt.savefig(root + "/Desktop/paper_fig/stm/con/diagnosis.png")
plt.show()
```
