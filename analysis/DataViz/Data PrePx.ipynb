# Document Topic Distribution

```python
import pandas as pd
import numpy as np

num_topics = 26
df = pd.read_csv(f"~/Desktop/R_js/data/gamma/pro{num_topics}_gamma.csv")
df['log_gamma'] = np.absolute(np.log(df.gamma))
display(df.head())
print(df.shape)
```

```python
lst = []
for t in range(1, num_topics+1): 
    df_topic = df[df.topic == t]
    lst.append(df_topic.gamma.sum())

tot = sum(lst)
lst = np.array([i/tot for i in lst])

print((lst).argsort()[::-1] + 1)


lst[::-1].sort()
print(lst)
print(sum(lst))
```

# Topic Term Distribution

```python
df = pd.read_csv(f"~/Desktop/R_js/data/beta/pro{num_topics}_beta.csv")
display(df.head())
tot_term = df[df.topic == 1]['term'].count()
print(tot_term)
```

```python
terms = df[df.topic == 1]["term"].values.tolist()
lst = []
i = 1
for t in terms: 
    term_prob = df[df.term == t]['beta'].sum()/num_topics
    lst.append([i, t, term_prob])
    i = i+1
df_term = pd.DataFrame(lst, columns=['word_id', 'word', 'word_prob'])
display(df_term.head())
```

```python


for t in terms: 
    term_prob = df_term.loc[df_term.word == t, "word_prob"].iloc[0]
    df.loc[df.term == t, "word_prob"] = term_prob
 
```

```python
lam = .5
df["term_freq"] = tot_term*df.word_prob
df["relevance"] = lam*df.beta + (1-lam)*df.beta/df.word_prob
```
