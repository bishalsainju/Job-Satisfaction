```python
import numpy as np
import pandas as pd
import seaborn as sns
import os
import matplotlib.pyplot as plt
import datetime 

data_path = os.path.join(os.path.expanduser('~'), 'Desktop/workspace/glassdoor/indeed/data')
df_company_list = pd.read_csv(data_path + '/scraper_data/review_site.csv')
```

```python
def merge_companies(companies_list=range(1, 51)):
    df_companies = pd.DataFrame()
    for company_index in companies_list:
        company_name = df_company_list.iloc[company_index - 1]['Company_Name']
        company_path = data_path + f'/companies/{company_index}_{company_name}/{company_name}.csv'
        df_company = pd.read_csv(company_path)
        df_companies = pd.concat([df_companies, df_company])
    return df_companies


def rating_barplot(df):
    grouped_df = df.groupby("Rating_Number")["Rating_Number"].count().reset_index(name="Count")
    display(grouped_df)
    sns.barplot(x="Rating_Number", y="Count", data=grouped_df).set_title("Frequency vs Rating")
    plt.show()
    

def overall_summary(companies_list=range(1, 51)):
    df = merge_companies(companies_list)
    display(f'Total no. of reviews: {df.shape}')
    rating_barplot(df)
    print(f"Count in each columns: ")
    display(df.count().reset_index())
    
    

# overall_summary([2, 3, 4, 5])
overall_summary()
```

```python
def company_summary(company_index):
    company_name = df_company_list.iloc[company_index - 1]['Company_Name']
    print(f'For {company_name}:')
    company_path = data_path + f'/companies/{company_index}_{company_name}/{company_name}.csv'
    df_company = pd.read_csv(company_path)
    display(df_company.head())
    display(df_company.shape)

company_summary(1)
```

```python
df['State'] = df['Reviewer_Job_Location'].str[-2:]

def filter_by_state(df, state):
    df_ = df[df.State == state]
    display(df_.head())
    display(df_.shape)
    return df_

df_state = filter_by_state(df, 'UT')
```

```python
df['Reviewed_Date'] = pd.to_datetime(df['Reviewed_Date'])
```

```python
def filter_by_date(df, start_date = (2018, 1, 1)):
    yr, mnth, dy = start_date
    df_ = df[df.Reviewed_Date > datetime.date(year=yr,month=mnth,day=dy)]
    display(df_.head())
    display(df_.shape)
    return df_

df_state = filter_by_date(df)
```
