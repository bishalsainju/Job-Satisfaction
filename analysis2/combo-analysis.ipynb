```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

num_topic = 20
data_path = f"~/Desktop/R_js/data1/estimates/combo-{num_topic}.csv"
df = pd.read_csv(data_path)
```

```python
df
```

```python


g = sns.FacetGrid(data=df)
# g.map(plt.scatter, 'x', 'y')

ax = g.axes[0,0]
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# ax.figure(figsize=(10, 10))
p1 = sns.regplot(x="Pro_Con", y="Job_Status", data=df, fit_reg=False, marker="+" ,
                scatter_kws={'s':400})
# add annotations one by one with a loop
for line in range(0,df.shape[0]):
     p1.text(df.Pro_Con[line]+.002, df.Job_Status[line], df.Topic_Num[line], 
             horizontalalignment='left', size='large', color='black', weight='bold')

plt.xlabel("Con_Pro")
plt.ylabel("Former_Current")
plt.xlim([-.15, .15])
plt.ylim([-.015, .02])
plt.gcf().set_size_inches(10, 10)
plt.show()
```
