```python
import pandas as pd
import numpy as np
```

```python
ff = pd.read_csv(f"~/Desktop/workspace/indeed/Job-Satisfaction/data/scraper_data/indeed_site50.csv")
ff.head()
```

```python
df = pd.DataFrame()

for index, row in ff.iterrows():
    company_rank = index+1
    company_name = row['Company_Name']
    df_company =  pd.read_csv(f"~/Desktop/workspace/indeed/Job-Satisfaction/data/companies/{company_rank}_{company_name}/"
                + f"{company_name}.csv")
    df = pd.concat([df, df_company], ignore_index=False)

df.head()
```
